# BÀI 1. LIỆT KÊ TỔ HỢP (TOHOP.*)

Cho tập hợp A có N phần tử đôi một khác nhau.
Mỗi phần tử là một ký tự từ A → Z.
Hãy liệt kê các tổ hợp chập K của N phần tử bằng thuật toán Backtracking.

```python
# Hàm Backtracking
def backtrack(start, N, K, A, current, result):
    if len(current) == K:
        result.append(current.copy())
        return

    for i in range(start, N):
        current.append(A[i])
        backtrack(i + 1, N, K, A, current, result)
        current.pop()
```

```python
# ===== Nhập dữ liệu =====
N, K = map(int, input().split())

A = []
for _ in range(N):
    A.append(input().strip())

# ===== Xử lý =====
result = []
current = []

backtrack(0, N, K, A, current, result)

# ===== Xuất kết quả =====
print(len(result))
for comb in result:
    print(" ".join(comb))
```

# BÀI 2. XẾP HẬU (XEPHAU.*)

Cho bàn cờ NxN. Hãy xếp N quân hậu sao cho không có quân nào ăn nhau.
In ra 1 nếu có cách xếp, 0 nếu không có cách xếp.
Nếu có, in ra một cách xếp dưới dạng ma trận 0/1

```python
# Kiểm tra vị trí đặt hậu
def is_safe(board, row, col, N):
    # kiểm tra cột
    for i in range(row):
        if board[i][col] == 1:
            return False

    # kiểm tra chéo trái
    i, j = row - 1, col - 1
    while i >= 0 and j >= 0:
        if board[i][j] == 1:
            return False
        i -= 1
        j -= 1

    # kiểm tra chéo phải
    i, j = row - 1, col + 1
    while i >= 0 and j < N:
        if board[i][j] == 1:
            return False
        i -= 1
        j += 1

    return True
```

```python
# Backtracking giải bài toán N-Queens
def solve_n_queens(board, row, N):
    if row == N:
        return True

    for col in range(N):
        if is_safe(board, row, col, N):
            board[row][col] = 1
            if solve_n_queens(board, row + 1, N):
                return True
            board[row][col] = 0

    return False
```

```python
# ===== Nhập dữ liệu =====
N = int(input())

board = [[0]*N for _ in range(N)]

# ===== Xử lý =====
if solve_n_queens(board, 0, N):
    print(1)
    for row in board:
        print(' '.join(map(str, row)))
else:
    print(0)
```

# BÀI 3. MÃ ĐI TUẦN (MADITUAN.*)

Trên bàn cờ NxN, con mã đặt tại (x0, y0).
Tìm lộ trình đi qua tất cả các ô đúng một lần và quay về vị trí ban đầu.
In 1 nếu có lộ trình, 0 nếu không có.

```python
# Các bước đi của quân mã
moves = [(2,1),(1,2),(-1,2),(-2,1),(-2,-1),(-1,-2),(1,-2),(2,-1)]

def is_valid(x, y, N, board):
    return 0 <= x < N and 0 <= y < N and board[x][y] == 0
```

```python
def solve_knight(board, x, y, move_count, N, start_x, start_y):
    if move_count == N*N:
        # kiểm tra quay về vị trí ban đầu
        for dx, dy in moves:
            if x + dx == start_x and y + dy == start_y:
                return True
        return False

    for dx, dy in moves:
        nx, ny = x + dx, y + dy
        if is_valid(nx, ny, N, board):
            board[nx][ny] = move_count + 1
            if solve_knight(board, nx, ny, move_count + 1, N, start_x, start_y):
                return True
            board[nx][ny] = 0

    return False
```

```python
# ===== Nhập dữ liệu =====
N, x0, y0 = map(int, input().split())

board = [[0]*N for _ in range(N)]
board[x0][y0] = 1

# ===== Xử lý =====
if solve_knight(board, x0, y0, 1, N, x0, y0):
    print(1)
    for row in board:
        print(' '.join(map(str, row)))
else:
    print(0)
```

# BÀI 4. TÔ MÀU ĐỒ THỊ (TOMAU.*)

Cho n thành phố và m cặp thành phố kề nhau.

Yêu cầu:
- Tô màu sao cho hai thành phố kề nhau không cùng màu.
- Số màu sử dụng là ít nhất có thể (dùng thuật toán tham lam).

Dữ liệu vào:
- Dòng đầu: n m
- m dòng tiếp theo: u v (hai thành phố kề nhau)

Dữ liệu ra:
- Dòng đầu: số màu k
- k dòng tiếp theo: các thành phố cùng một màu

```python


def is_safe(v, graph, color, c):
    for u in graph[v]:
        if color[u] == c:
            return False
    return True


def backtrack(v, n, graph, color, m):
    if v == n:
        return True

    for c in range(1, m + 1):
        if is_safe(v, graph, color, c):
            color[v] = c

            if backtrack(v + 1, n, graph, color, m):
                return True

            color[v] = 0

    return False
```

```python
# Nhap du lieu

n = int(input())
graph = [[] for _ in range(n)]

while True:
    try:
        u, v = map(int, input().split())
        graph[u-1].append(v-1)
        graph[v-1].append(u-1)
    except:
        break
```

```python
color = [0] * n

# Tim so mau it nhat
for m in range(1, n + 1):
    if backtrack(0, n, graph, color, m):
        break
```

```python
print(m)

groups = {}

for i in range(n):
    if color[i] not in groups:
        groups[color[i]] = []

    groups[color[i]].append(i + 1)

for g in groups.values():
    print(*g)
```

# BÀI 5: LIỆT KÊ HOÁN VỊ (Backtracking)

Cho tập hợp A có N phần tử đôi một khác nhau.
Mỗi phần tử là một ký tự từ A → Z.

Yêu cầu:
- Liệt kê tất cả các hoán vị của tập hợp A.

Ý tưởng thuật toán:
- Dùng Backtracking để chọn từng phần tử.
- Mỗi bước chọn một phần tử chưa dùng.
- Khi chọn đủ N phần tử thì lưu hoán vị

```python


def backtrack(A, used, current, result):

    if len(current) == len(A):
        result.append(current.copy())
        return

    for i in range(len(A)):
        if not used[i]:
            used[i] = True
            current.append(A[i])

            backtrack(A, used, current, result)

            current.pop()
            used[i] = False
```

```python
# Nhap du lieu

N = int(input())

A = []
for _ in range(N):
    A.append(input().strip())
```

```python
result = []
current = []
used = [False] * N

backtrack(A, used, current, result)
```

```python
print(len(result))

for perm in result:
    print(" ".join(perm))
```

# BÀI 2: ĐIỀN BẢNG SỐ (FUJIWARA)

Trong lưới 9×9 cần điền các số từ 1 → 9 sao cho:
- Mỗi hàng chứa đủ các số 1 → 9.
- Mỗi cột chứa đủ các số 1 → 9.
- Mỗi ô vuông 3×3 chứa đủ các số 1 → 9.

Dữ liệu:
- 9 dòng, mỗi dòng 9 ký tự.
- '.' nghĩa là ô chưa có số.

Kết quả:
- In ra bảng Sudoku hoàn chỉnh
- Nếu không giải được in: IMPOSSIBLE

```python
# Kiem tra so co the dat vao o (r,c) hay khong

def is_safe(board, r, c, num):

    # kiem tra hang
    for i in range(9):
        if board[r][i] == num:
            return False

    # kiem tra cot
    for i in range(9):
        if board[i][c] == num:
            return False

    # kiem tra o 3x3
    start_r = (r // 3) * 3
    start_c = (c // 3) * 3

    for i in range(start_r, start_r + 3):
        for j in range(start_c, start_c + 3):
            if board[i][j] == num:
                return False

    return True
```

```python
def solve(board):

    for r in range(9):
        for c in range(9):

            if board[r][c] == '.':

                for num in '123456789':

                    if is_safe(board, r, c, num):

                        board[r][c] = num

                        if solve(board):
                            return True

                        board[r][c] = '.'

                return False

    return True
```

```python
# Nhap du lieu

board = []

for _ in range(9):
    row = list(input().strip())
    board.append(row)
```

```python
if solve(board):

    for row in board:
        print("".join(row))

else:
    print("IMPOSSIBLE")
```
